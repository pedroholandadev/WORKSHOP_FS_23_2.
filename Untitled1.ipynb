```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dados.csv')
```

```python
print(df)
```

```python
df.head()
```

```python
# Não consegui achar os numeros máximos, média e menor número usando códigos.
df.head().max()
df.head().mean()
df.head().min()
```

```python
# Agrupamentos dos alunos por etnia e suas respectivas médias em Matemática
df2 = df.groupby('race/ethnicity')[['math score']].mean()
df2
```

```python
# Agrupamento dos alunos por Nível de escolaridade dos pais e suas respectivas notas em Escrita
df3 = df.groupby('parental level of education')[['writing score']].mean()
df3
```

```python
# Agrupamento dos alunos por Participação em cursos preparatórios para o teste e suas respectivas médias em Leitura
df4 = df.groupby('test preparation course')[['reading score']].mean()
df4
```

```python
df2.plot(kind = 'bar')
```

```python
df3.plot(kind = 'pie', subplots = True)
```

```python
df4.plot(kind = 'barh')
```
